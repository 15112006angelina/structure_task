```python
students = []

n = int(input("Введіть кількість студентів: "))

for i in range(n):
    pip = input("Введіть П.І.П.: ")
    birth_date = input("Введіть дату народження (рррр-мм-дд): ")
    address = input("Введіть адресу: ")
    course = int(input("Введіть курс: "))
    group = input("Введіть групу: ")

    student = {
        "ПІП": pip,
        "дата_народження": birth_date,
        "адреса": address,
        "курс": course,
        "група": group
    }

    students.append(student)

print("\nПрізвища студентів 2-го курсу:")
for student in students:
    if student["курс"] == 2:
        прізвище = student["ПІП"].split()[0]
        print(прізвище)
```

```python
class Student:
    def __init__(self, pip, birth_date, address, course, group):
        self.pip = pip
        self.birth_date = birth_date
        self.address = address
        self.course = course
        self.group = group

students = []

n = int(input("Введіть кількість студентів: "))

for i in range(n):
    pip = input("Введіть П.І.П.: ")
    birth_date = input("Введіть дату народження (рррр-мм-дд): ")
    address = input("Введіть адресу: ")
    course = int(input("Введіть курс: "))
    group = input("Введіть групу: ")

    student = Student(pip, birth_date, address, course, group)
    students.append(student)

print("\nПрізвища студентів 2-го курсу:")
for student in students:
    if student.course == 2:
        прізвище = student.pip.split()[0]
        print(прізвище)
```

```python
cars = [
    {"марка": "Toyota Corolla", "рік_випуску": 2020, "ціна": 35000},
    {"марка": "Honda Civic", "рік_випуску": 2019, "ціна": 42000},
    {"марка": "Skoda Octavia", "рік_випуску": 2018, "ціна": 38000}
]

print("Наявність на складі:")
for car in cars:
    print(car)

print("\nАвтомобілі дешевші за 40000 грн:")
for car in cars:
    if car["ціна"] < 40000:
        print(f'{car["марка"]}, {car["рік_випуску"]}')
```

```python
class Car:
    def __init__(self, brand, year, price):
        self.brand = brand
        self.year = year
        self.price = price

cars = [
    Car("Toyota Corolla", 2020, 35000),
    Car("Honda Civic", 2019, 42000),
    Car("Skoda Octavia", 2018, 38000)
]

print("Наявність на складі:")
for car in cars:
    print(f"{car.brand}, {car.year}, {car.price} грн")

print("\nАвтомобілі дешевші за 40000 грн:")
for car in cars:
    if car.price < 40000:
        print(f"{car.brand}, {car.year}")
```
